```python
x = np.array([2,4,6,8])
y = np.array([10,20,30,40])
z = np.array([12,14,16,18])
a = np.array([10,12,16,18])
```

```python
plt.plot(x,y)
```

```python
plt.scatter(x,y)
```

```python
plt.plot(x,y)
plt.plot(x,z)
plt.plot(y,z)
plt.plot(a,z)
plt.xlabel("age in months")
plt.ylabel("weight")
plt.title("Graph of age and weight")
plt.legend(["x","y","line of z","a"])
plt.grid(True)
plt.show()
```

```python
plt.bar(x,y,width = .5)
plt.show()
```
