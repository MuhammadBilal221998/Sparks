```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.cluster import KMeans
```

```python
#Scatter Plot

plt.scatter(user_data.Discount,user_data.Shipping_Amount)
plt.xlabel("Discount")
plt.ylabel("Shipping_Amount")
plt.title("Scatter plot of Discount and Shipping_Amount")
plt.show()
```

```python
#Clustering

user_data1 = pd.read_csv(r"C:\Users\MuhammadBilal\OneDrive - Unison\Desktop\Logisticregression123.csv")

cluster = KMeans(4)
a = cluster.fit_predict(user_data1)
a

user_data1["cluster"] = a
user_data1
```

```python
# Visualization

plt.scatter(user_data1.Status, user_data1.Order_ID, c = user_data1.cluster, cmap = 'rainbow')
plt.xlabel("Discount")
plt.ylabel("Shipping_Amount")
plt.title("Scatter plot")
plt.show()
```
