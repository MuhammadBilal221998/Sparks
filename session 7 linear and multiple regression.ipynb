```python
from sklearn.linear_model import LinearRegression
import numpy as np
import statsmodels.api as sm
```

```python
x = user_data.Shipping_Amount
y = user_data.Discount

x = sm.add_constant(x) #this command shows the value of constant

model = sm.OLS(y,x)  ## Ordinary Least Square
results = model.fit() ## Draws line from 3 dots
results.summary()
```

```python
x = user_data[['Order_Quantity','Unit_Sell_Price']]
y = user_data.Discount

x = sm.add_constant(x)

model = sm.OLS(y,x)
results = model.fit()
results.summary()
```
